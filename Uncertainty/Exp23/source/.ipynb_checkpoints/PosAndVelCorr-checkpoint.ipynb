Deeper dive into the position and velocity correlations in EPG cells

```python
#Import all the modules we will be using

import os #import module to view and manipulate directories
import scipy.io as sio #import module to open matlab files
import matplotlib.pyplot as plt #import the plotting module
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
from scipy.stats import circstd, pearsonr
```

```python
#Load data for single analyses

mainDir = os.getcwd()[:-7]
folders = os.listdir(mainDir + '\data')  #get the list of folders in it
flyNum = len(folders) #determine fly number
```

```python
# Load group data

groupData = sio.loadmat(mainDir + '\data\\allDarkData.mat', struct_as_record=True)
flyGroupData = groupData['Data']

#To access the fields in flyGroupData, use flyGroupData.dtype
```

```python
#Analyze the correlation between phase value and fly position

#1) Using the Seelig 2015 method
#Pearson's correlation coefficients were computed between two entire 'unwrapped' time series

#Looking at  the uncertain part of the trial
fig,(ax1, ax2, ax3, ax4) = plt.subplots(4,1, figsize = (15, 10), dpi=100)

actualHeading1 = np.unwrap(flyGroupData[0,0]['panel_angle'][0,np.asscalar(flyGroupData[0,0]['StartDarkness']):np.asscalar(flyGroupData[0,0]['EndDarkness'])])
decodedHeading1 = np.unwrap(flyGroupData[0,0]['phase_value'][0,np.asscalar(flyGroupData[0,0]['StartDarkness']):np.asscalar(flyGroupData[0,0]['EndDarkness'])])

ax1.plot(flyGroupData[0,0]['panel_angle'][0,np.asscalar(flyGroupData[0,0]['StartDarkness']):np.asscalar(flyGroupData[0,0]['EndDarkness'])])
ax1.plot(flyGroupData[0,0]['phase_value'][0,np.asscalar(flyGroupData[0,0]['StartDarkness']):np.asscalar(flyGroupData[0,0]['EndDarkness'])])
ax1.set(ylabel='Rad')
ax1.legend(('Fly heading', 'EPG phase'),loc='upper right', frameon='True', shadow = 'True')
ax1.set(title='Example Uncertain state in fly 1')

ax2.plot(flyGroupData[0,0]['offset'][0,np.asscalar(flyGroupData[0,0]['StartDarkness']):np.asscalar(flyGroupData[0,0]['EndDarkness'])],color ='black',linestyle=':')
ax2.set(ylabel='Offset (rad)')

ax3.plot(actualHeading1)
ax3.plot(decodedHeading1)
ax3.set(ylabel='unwrapped rad')
ax3.set(xlabel='time')
ax3.legend(('Fly heading', 'EPG phase'),loc='upper right', frameon='True', shadow = 'True')

ax4.scatter(actualHeading1,decodedHeading1)
ax4.set(title='correlation between unwrapped EPG and unwrapped actual heading')
#get correlation
dataCorr = pearsonr(actualHeading1,decodedHeading1)
ax4.text(1,8,'Pearson corr = ' + str(round(dataCorr[0],2)), fontsize=16)

fig.tight_layout(pad=3.0)
plt.savefig(mainDir + "\plots\PosCorrelationEx.pdf", dpi=150)

#This is pretty strange. Even though the offset appears super variable, when we unwrap both time series, they actually look like they correlate a lot.
```

```python
#Re-calculate the offset to see what I get

#define circular distance function
def circ_dist(x,y):
    return np.angle(np.exp(1j*x)/np.exp(1j*y))

offset = -circ_dist(actualHeading1,decodedHeading1)

fig,ax1 = plt.subplots(1,1, figsize = (15, 10), dpi=100)
ax1.plot(offset)
ax1.plot(flyGroupData[0,0]['offset'][0,np.asscalar(flyGroupData[0,0]['StartDarkness']):np.asscalar(flyGroupData[0,0]['EndDarkness'])],color ='black',linestyle=':')
ax1.set(ylabel='offset (rad)')
ax1.set(xlabel='Time')
ax1.legend(('Newly calculated offset', 'Previously calculated offset'),loc='upper right', frameon='True', shadow = 'True')

#I get the same thing! So the offset is super variable, but the correlation of unwrapped data is almost perfect, how is that possible?
```

```python
#An example of the opposite appears to be the last certain state for fly 0, where the offset is very stable but the corr is much lower

fig,(ax1, ax2, ax3, ax4) = plt.subplots(4,1, figsize = (15, 10), dpi=100)

actualHeading2 = np.unwrap(flyGroupData[0,0]['panel_angle'][0,np.asscalar(flyGroupData[0,0]['EndDarkness']+1):-1])
decodedHeading2 = np.unwrap(flyGroupData[0,0]['phase_value'][0,np.asscalar(flyGroupData[0,0]['EndDarkness']+1):-1])

fig.tight_layout(h_pad=3)

ax1.plot(flyGroupData[0,0]['panel_angle'][0,np.asscalar(flyGroupData[0,0]['EndDarkness']+1):-1])
ax1.plot(flyGroupData[0,0]['phase_value'][0,np.asscalar(flyGroupData[0,0]['EndDarkness']+1):-1])
ax1.set(ylabel='Rad')
ax1.legend(('Fly heading', 'EPG phase'),loc='upper right', frameon='True', shadow = 'True')

ax2.plot(flyGroupData[0,0]['offset'][0,np.asscalar(flyGroupData[0,0]['EndDarkness']+1):-1],color ='black',linestyle=':')
ax2.set(ylabel='Offset (rad)')

ax3.plot(actualHeading2)
ax3.plot(decodedHeading2)
ax3.set(ylabel='unwrapped rad')
ax3.set(xlabel='time')
ax3.legend(('Fly heading', 'EPG phase'),loc='upper right', frameon='True', shadow = 'True')

ax4.scatter(actualHeading2,decodedHeading2)
ax4.set(title='correlation between EPG and actual heading')
#get correlation
dataCorr = pearsonr(actualHeading2,decodedHeading2)
ax4.text(0,0,'Pearson corr = ' + str(round(dataCorr[0],2)), fontsize=16)

plt.savefig(mainDir + "\plots\PosCorrelationEx2.png", dpi=150)
```

```python
#2) Using the Maimon lab criteria from Green et al 2017
#His correlations overall seem much lower (this might be because of trial duration, method for taking the corr, etc)

#for closed-loop bar experiments, he computes the circular correlation between the GCaMP phase and bar position
#he calculates these correlations for different time lags between the phase and ball signals, and reports the higher correlation.
#for all phase correlations, he only includes data during which the fly is walking with a speed of at least 1 mm/s
#he also requires that the peak activity is greater than 0.8 DF/F to ensure that the phase is properly estimated.
from astropy.stats import circcorrcoef
circcorrcoef(actualHeading1,decodedHeading1)
circcorrcoef(actualHeading2,decodedHeading2)

#try different lags
#Initialize a vector of all the possible lags
lags = np.arange(-30, 31, 1)
circCoef1 = np.zeros((len(lags),1))
circCoef2 = np.zeros((len(lags),1))
#run a loop for every element in the lags vector
for lag in lags:
    #circularly shift the decoded heading by that lag
    shiftedDecodHeading1 = np.roll(decodedHeading1,lags[lag])
    shiftedDecodHeading2 = np.roll(decodedHeading2,lags[lag])
    #compute the circular correlation and save
    circCoef1[lag] = circcorrcoef(actualHeading1,shiftedDecodHeading1)    
    circCoef2[lag] = circcorrcoef(actualHeading2,shiftedDecodHeading2)
    
#plot the circular correlation coefficient for the different lags
fig, (ax1,ax2) = plt.subplots(1,2,figsize = (10, 10))
ax1.plot(lags,circCoef1,marker='o')
ax1.set_ylabel('Circular correlation for position', fontsize=14)
ax1.set_xlabel('Lag', fontsize=14)
ax1.set_title('Uncertain state', fontsize=16)
ax1.set_ylim(-1,1)

ax2.plot(lags,circCoef2,marker='o')
ax2.set_xlabel('Lag', fontsize=14)
ax2.set_title('Certain state', fontsize=16)
ax2.set_ylim(-1,1)

plt.savefig(mainDir + "\plots\CircCorrAllData.png", dpi=150)
#these values make a lot more sense with the offset variation! but I don't understand why they are so different from the Pearson correlation of the unwrapped data.
```

```python
#Using velocity instead of position
#for experiments in the dark, he computes the Pearson correlation between GCaMP phase velocity and ball velocity. This is pretty low for both
pearsonr(np.ediff1d(actualHeading1),np.ediff1d(decodedHeading1))[0]
pearsonr(np.ediff1d(actualHeading2),np.ediff1d(decodedHeading2))[0]
#The pearson correlation between them is pretty terrible

#try different lags
#Initialize a vector of all the possible lags
lags = np.arange(-30, 31, 1)
pearsonCoef1 = np.zeros((len(lags),1))
pearsonCoef2 = np.zeros((len(lags),1))
#run a loop for every element in the lags vector
for lag in lags:
    #circularly shift the decoded heading by that lag
    shiftedDecodHeading1 = np.roll(np.ediff1d(decodedHeading1),lags[lag])
    shiftedDecodHeading2 = np.roll(np.ediff1d(decodedHeading2),lags[lag])
    #compute the circular correlation and save
    pearsonCoef1[lag] = pearsonr(np.ediff1d(actualHeading1),shiftedDecodHeading1)[0]    
    pearsonCoef2[lag] = pearsonr(np.ediff1d(actualHeading2),shiftedDecodHeading2)[0]
    
#plot the circular correlation coefficient for the different lags
fig, (ax1,ax2) = plt.subplots(1,2,figsize = (10, 10))
ax1.plot(lags,pearsonCoef1,marker='o')
ax1.set_ylabel('Pearson correlation for velocity',fontsize=14)
ax1.set_xlabel('Lag',fontsize=14)
ax1.set_title('Uncertain state',fontsize=16)
ax1.set_ylim(-1,1)

ax2.plot(lags,pearsonCoef2,marker='o')
ax2.set(xlabel='Lag')
ax2.set_ylim(-1,1)
ax2.set_title('Certain state',fontsize=16)
fig.tight_layout(pad=3.0)

plt.savefig(mainDir + "\plots\PearsonCorrAllDataVel.png", dpi=150)
```

```python
test = UsefulUncertainHeading[~np.isnan(UsefulUncertainHeading)] 
test2 = UsefulUncertainDecHeading[~np.isnan(UsefulUncertainDecHeading)] 

test3 = UsefulCertainHeading[~np.isnan(UsefulCertainHeading)] 
test4 = UsefulCertainDecHeading[~np.isnan(UsefulCertainDecHeading)] 


#try different lags
#Initialize a vector of all the possible lags
lags = np.arange(-30, 31, 1)
circCoef1 = np.zeros((len(lags),1))
circCoef2 = np.zeros((len(lags),1))
#run a loop for every element in the lags vector
for lag in lags:
    #circularly shift the decoded heading by that lag
    shiftedDecodHeading1 = np.roll(test2,lags[lag])
    shiftedDecodHeading2 = np.roll(test4,lags[lag])   
    #compute the circular correlation and save
    circCoef1[lag] = circcorrcoef(test,shiftedDecodHeading1)    
    circCoef2[lag] = circcorrcoef(test3,shiftedDecodHeading2)    
    
#plot the circular correlation coefficient for the different lags
fig, (ax1,ax2) = plt.subplots(1,2,figsize = (10, 10))
ax1.plot(lags,circCoef1,marker='o')
ax1.set_ylabel('Circular correlation for position',fontsize=14)
ax1.set_xlabel('Lag',fontsize=14)
ax1.set_title('Uncertain state',fontsize=16)
ax1.set_ylim(-1,1)

ax2.plot(lags,circCoef2,marker='o')
ax2.set_xlabel('Lag',fontsize=14)
ax2.set_title('Certain state',fontsize=16)
ax2.set_ylim(-1,1)

plt.savefig(mainDir + "\plots\CircCorrPosFilteredData.png", dpi=150)
```
